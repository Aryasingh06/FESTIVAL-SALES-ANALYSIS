```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('Festival Sales Data.csv',encoding='unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
#DATA CLEANING
# drop unrelated/blank column
df.drop(['Status','unnamed1'],axis=1 ,inplace=True)
```

```python
# count number of null values 
pd.isnull(df).sum()
```

```python
#drop null values
df.dropna(inplace=True)
```

```python
df.shape
```

```python
#change datatype
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

```python
df.columns
```

```python
#rename column
df.rename(columns= {'Marital_Status':'Marital_Condition'})
```

```python
df.describe()
```

## Expolatory Data Analysis

```python
# GENDER
# COUNT WHICH GENDER PURCHASES MORE IN FESTIVAL SEASON (GENDER WISE)

ax = sns.countplot(x = 'Gender',data = df)
plt.savefig("images/gender_analysis.png")
```

```python
# plotting a bar graph for gender vs amount which gender generate more sales.
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
                                               #“Group data by Gender → sum Amount → sort highest to lowest”

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
plt.savefig("images/gendervsamount.png")
```

###### FROM ABOVE GRAPH WE CAN SEE THAT MOST OF THE BUYERS ARE FEMALES AND EVEN PURCHASING POWER OF FEMALE GREATER THAN MALES

```python
#AGE
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
plt.savefig("images/agegrp.png")
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

#### most of buyers are between age group 26-35 yrs females.

```python
#STATES
df.columns
```

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
plt.savefig("images/statevsorders.png")
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
plt.savefig("images/statevsamount.png")
```

### Most orders and amount generated by UTTAR PRADESH. 

```python
#maritial status
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(7,5)})
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

### Most buyers are married woman and having purchasing power  high

```python
# occupation
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')
plt.savefig("images/maritial_status.png")
```

```python
# which occupation people  spent more amount
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
plt.savefig("images/occupationvsamount.png")
```

### From the above graph we can see that most of the buyers are woring in IT sector,Healthcare.

```python
#product category
sns.set(rc={'figure.figsize':(25,10)})
ax = sns.countplot(data = df, x = 'Product_Category')
plt.savefig("images/productcount.png")
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
plt.savefig("images/productamount.png")
```

### FROM ABOVE GRAPH WE CAN SEE THAT MOST OF THE SOLD PRODUCTS ARE FROM FOOD,CLOTHING AND ELECTRONICS CATEGORY.

## CONCLUSION
Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
